```python
# import packages
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans

# generate data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=1)

# k-means clustering
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=1)
pred_y = kmeans.fit_predict(X)

# visualization
xmin = -15
xmax = 5

ymin = -11
ymax = 7

ax1 = plt.subplot(121)
ax1.scatter(X[:,0], X[:,1])
ax1.set_xlim([xmin,xmax])
ax1.set_ylim([ymin,ymax])

ax2 = plt.subplot(122)
ax2.scatter(X[:,0], X[:,1], c=pred_y)
ax2.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=50, c='red')
ax2.set_xlim([xmin,xmax])
ax2.set_ylim([ymin,ymax])

plt.show()
```
